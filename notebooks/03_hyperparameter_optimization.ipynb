# Hyperparameter Optimization

Hyperparameters were tuned using Optuna with three-fold cross-validation to maximize the validation Macro F1 score.

```python
import optuna
import pandas as pd
from pathlib import Path

from optuna.visualization import (
    plot_optimization_history,
    plot_param_importances,
)

REPO_ROOT = Path().resolve().parent

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
```

```python
storage = "sqlite:///" + str(REPO_ROOT / "artifacts" / "optimization" / "optuna.db")

study_initial = optuna.load_study(
    study_name="journaling_downsampled",
    storage=storage,
)

study_refined = optuna.load_study(
    study_name="journaling_downsampled_narrowed",
    storage=storage,
)
```

## Best Configuration

The best trial from the refined search was selected as the final training configuration.

```python
best_initial = study_initial.best_trial.params.copy()
best_refined = study_refined.best_trial.params

final_params = best_initial | best_refined
summary_df = pd.DataFrame(
    final_params.items(),
    columns=["Hyperparameter", "Value"],
)

summary_df
```

## Optimization History

The optimization history shows how the best validation score improved over successive trials.

```python
plot_optimization_history(study_initial)
```
